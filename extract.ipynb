```python
from bs4 import BeautifulSoup
import requests

url = 'https://www.news-medical.net/'
article_links = []

html_text = requests.get(url).text

soup = BeautifulSoup(html_text, 'html.parser')

all_links = soup.find_all('a')

links_ones = soup.find_all('article')
print(len(links_ones))

for tag in links_ones:
    article_link = url + tag.a['href']
    # print(article_link)
    article_links.append(article_link)
    # print(tag.a['href'])
    # print(tag['href'])

test_link = article_links[0]

new_html = requests.get(test_link).text
new_soup = BeautifulSoup(new_html, 'html.parser')

all_paragraphs = new_soup.find_all('p')

articles_final = []

para1 = ''

for para in all_paragraphs:
    para1 += para.text.strip().replace('\n', ' ')
    para1 += '\n'

articles_final.append(para1)
with open('news-medical.txt', 'w') as f:
    f.write(para1)
print(articles_final)
# print(all_paragraphs)
# print(new_soup.prettify())

# print(links_ones)

# print(soup.prettify())
# with open('worldbank.txt', 'w') as f:
#     f.write(soup.prettify())
```
